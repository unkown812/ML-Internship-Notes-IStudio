# Absolute Maximum Scaling:

```python
import pandas as pd
```

```python
df=pd.DataFrame({"LoArea":[8450,9600,11250,9550,14260],"MSSubCLass":[60,20,60,70,60]})
```

```python
df
```

```python
#finding the absolute value:

import numpy as np

max_val=np.max(np.abs(df))
max_val
```

```python
print((df-max_val)/max_val)
```

# Min Max Scaling:

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()

scaled_data=scaler.fit_transform(df)

scaled_df=pd.DataFrame(scaled_data,columns=df.columns)
```

```python
scaled_df
```

```python
# this approach is also sensitive towards the outliers.
```

# Normalization:

```python
from sklearn.preprocessing import Normalizer

scaler=Normalizer()
scaled_entry=scaler.fit_transform(df)
scaled_norm=pd.DataFrame(scaled_entry,columns=df.columns)
print(scaled_norm)
```

# Standardization:

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
scale.fit_transform(df)
```
